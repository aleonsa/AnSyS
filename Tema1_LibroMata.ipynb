```python
import numpy as np
import matplotlib as mlp
import matplotlib.pyplot as plt
import seaborn as sns

from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
t = np.linspace(0,3,6000)
x1 = np.sin(2*np.pi*t)
```

```python
plt.plot(t,x1)
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title(r'Plot of signal $x(t)= \sin(2\pi t)$');

sns.set_style("whitegrid")
plt.show()
```

```python
# Figura 1.8 Sistema invariante en el tiempo
t = np.linspace(0,3,600)
x = np.sin(2*np.pi*t)
r = 10
```

```python
# funcin invariante en el tiempo
y1 = x/r
plt.plot(t,y1)
plt.axis([0,3,-0.11,0.11])
plt.xlabel('t[s]')
plt.show()
```

```python
# Calculando la funcin desplazada
t = np.linspace(0,3,600)
x = [0]*100
x.extend(np.sin(2*np.pi*np.linspace(0,2.5,500)))

len(x)
```

```python
y2 = []
for i in x: y2.append(i/r)

plt.plot(t,y2)
plt.axis([0,3,-0.11,0.11])
plt.xlabel('t[s]')
plt.show()
```

```python

def f(a):
    t = np.linspace(0,3,6000)
    x1 = a*np.sin(2*np.pi*t)
    x2 = a*np.cos(2*np.pi*t)
    
    fig, axs = plt.subplots(2, sharey=True)
    axs[0].plot(t,x1)
    axs[1].plot(t,x2, 'tab:orange')
    
    plt.ylim(-5,5)
    plt.show()
    
    
interact(f,a=(0,5,0.2)) 
```
